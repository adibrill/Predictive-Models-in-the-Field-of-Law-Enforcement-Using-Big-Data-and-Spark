#### 1.5.1 T-test

The t-test is a statistical test used to compare the means of two samples and determine whether they are significantly different from each other. Specifically, the t-test compares the difference between the means of the two samples to the variability within the samples.

The t-test is a parametric test, which means that it assumes that the data are normally distributed and have equal variances in the two samples. There are two types of t-tests: the independent samples t-test and the paired samples t-test.

The independent samples t-test is used when the two samples are independent of each other, meaning that the observations in one sample are not related to the observations in the other sample. 

The null hypothesis of the t-test is a statement of equality between the means of two samples. Specifically, the null hypothesis states that there is no significant difference between the means of the two samples, and any observed distinction is due to chance or random sampling variability.

If the p-value of the t-test is greater than the significance level (0.05) than we can say that it is likely that there is no significant difference between the means of the two samples.

#### 1.5.2 Chi-Square Test

The chi-square test is a statistical test used to determine whether there is a significant association between two categorical variables. The test is based on the chi-square distribution and can be used for both goodness-of-fit tests and tests of independence.

To perform the chi-square test, a contingency table is constructed that displays the observed frequencies of each category for the two variables. The expected frequencies are then calculated based on the null hypothesis, using the marginal totals of the table. The chi-square statistic is then calculated by summing the squared difference between the observed and expected frequencies, divided by the expected frequency, across all categories.

The null hypothesis of the chi-square test is the observed frequencies in each category are not significantly different from the expected frequencies.

If the p-value of the chi-square test is more than 0.05, it means that there is not enough evidence to reject the null hypothesis that there is no significant association or difference between the categorical variables being tested. In other words, the observed data does not provide enough evidence to conclude that there is no causation or association between the variables.



#### 1.5.3 Augmented Dickeyâ€“Fuller Test

The Augmented Dickey-Fuller (ADF) test is a statistical test used to determine whether a time series is stationary or non-stationary. Stationarity refers to a time series having a constant mean, constant variance, and no correlation between the observations at different time points. Non-stationary time series have trends, seasonality, or other patterns that change over time.

The ADF test is an extension of the Dickey-Fuller test, which tests the null hypothesis that a time series has a unit root. The ADF test includes additional lagged terms in the regression equation to account for serial correlation in the data.

The test involves regressing the time series on its lagged values and testing whether the coefficient on the lagged series is significantly different from one. If the coefficient is significantly different from one, then the null hypothesis of a unit root is rejected, and the time series is considered stationary. If the coefficient is not significantly different from one, then the null hypothesis cannot be rejected, and the time series is considered non-stationary.

The ADF test is often used in time series analysis to determine whether a time series needs to be transformed or differenced to make it stationary. Stationary time series are easier to model and analyze, and many time series analysis techniques, such as ARIMA models, assume stationarity.

It is important to note that the ADF test has limitations and assumptions. For instance, the test assumes that the time series is stationary around a deterministic trend, and it may not work well for time series with structural breaks or other non-linear patterns. Therefore, it is important to carefully interpret the results of the ADF test and consider other factors, such as the visual inspection of the time series and the context of the data.

Null Hypothesis: The data is not stationary.

Alternative Hypothesis: The data is stationary.

For the data to be stationary (ie. reject the null hypothesis), the ADF test should have a p-value greater than the significance level.

The next step is combining essentially identical offenses to lower the categories quantity and improve model probability.

```python
def GroupDfbyPrimaryTypes(datadf):
    '''
    This function helps combine PRIMARY CRIME TYPES
    '''
    datadf = datadf.withColumn("Primary Type", \
              when(datadf["Primary Type"] == 'OTHER NARCOTIC VIOLATION','NARCOTICS').otherwise(datadf['Primary Type']))
    datadf = datadf.withColumn("Primary Type", \
              when((datadf["Primary Type"] == 'PROSTITUTION') | (datadf["Primary Type"] == 'CRIM SEXUAL ASSAULT'),'SEX OFFENSE').otherwise(datadf['Primary Type']))    
    datadf = datadf.withColumn("Primary Type", \
              when((datadf["Primary Type"] == 'LIQUOR LAW VIOLATION') | (datadf["Primary Type"] == 'RITUALISM') | (datadf["Primary Type"] == 'GAMBLING'),'OTHER OFFENSE').otherwise(datadf['Primary Type']))
    datadf = datadf.withColumn("Primary Type", \
              when((datadf["Primary Type"] == 'CRIMINAL TRESPASS') | (datadf["Primary Type"] == 'ROBBERY'),'ROBBARY or TRESPASS').otherwise(datadf['Primary Type']))
    datadf = datadf.withColumn("Primary Type", \
              when((datadf["Primary Type"] == 'INTERFERENCE WITH PUBLIC OFFICER') ,'PUBLIC PEACE VIOLATION').otherwise(datadf['Primary Type']))
    datadf = datadf.withColumn("Primary Type", \
              when((datadf["Primary Type"] == 'INTIMIDATION') | (datadf["Primary Type"] == 'STALKING'),'INTIMIDATION or STALKING').otherwise(datadf['Primary Type']))
    return datadf
```

# 2.2 Crime Classification

## 2.2.1 Classifying and Predicting Arrest Occurance

This code first loads the Chicago crimes relevant columns. It then splits the data into training and testing sets, and a logistic regression model is trained on the training set using cross-validation. Afterward, the model is evaluated on the test set using the BinaryClassificationEvaluator. The evaluator computes the area under the ROC curve (AUC) and finally, presents the test set accuracy.


```python
CrimesDataset = CrimesDataset.withColumn("Latitude",CrimesDataset.Latitude.cast('double'))
CrimesDataset = CrimesDataset.withColumn("Longitude",CrimesDataset.Longitude.cast('double'))
CrimesDataset = CrimesDataset.withColumn("Beat",CrimesDataset.Beat.cast(IntegerType()))
CrimesDataset = CrimesDataset.withColumn("District",CrimesDataset.District.cast(IntegerType()))
CrimesDataset = CrimesDataset.withColumn("Ward",CrimesDataset.Ward.cast(IntegerType()))
CrimesDataset = CrimesDataset.withColumn("Day_Of_Week",CrimesDataset.Day_Of_Week.cast(IntegerType()))
CrimesDataset = CrimesDataset.withColumn("Month_Of_Year",CrimesDataset.Month_Of_Year.cast(IntegerType()))
CrimesDataset = CrimesDataset.withColumn("Year",CrimesDataset.Year.cast(IntegerType()))
CrimesDataset = CrimesDataset.withColumn("Arrest",CrimesDataset.Arrest.cast(IntegerType()))
```

```python
# select only the columns relevant for the model

df = CrimesDataset.select("Beat","District","Ward","Month_Of_Year", "Arrest")

# limit the model to 150 thousand records to improve code runtime
#df = df.limit(150000)

# change the arrest column name to "label"
df = df.withColumnRenamed("Arrest","label")


cat_columns = ["Beat","District","Ward", "Month_Of_Year"]

# Combine all features into a single feature vector
assembler = VectorAssembler(inputCols=cat_columns, outputCol="features")
df = assembler.transform(df)

# Split the dataset into training and testing sets
(trainingData, testData) = df.randomSplit([0.7, 0.3], seed=3255)

# Train the logistic regression model
lr = LogisticRegression(labelCol="label", featuresCol="features", maxIter=10, regParam=0.3, elasticNetParam=0.8)

# Define the grid of hyperparameters to tune
paramGrid = ParamGridBuilder() \
    .addGrid(lr.regParam, [0.1, 0.3, 0.5]) \
    .addGrid(lr.elasticNetParam, [0.0, 0.5, 1.0]) \
    .build()

```

```python
# Use cross-validation to select the best hyperparameters
evaluator = BinaryClassificationEvaluator()
cv = CrossValidator(estimator=lr, estimatorParamMaps=paramGrid, evaluator=evaluator, numFolds=5)
cvModel = cv.fit(trainingData)

# Evaluate the model on the test set
predictions_arrest = cvModel.transform(testData)
accuracy_arrest = evaluator.evaluate(predictions_arrest)
print("Test set accuracy = {:.7f}".format(accuracy_arrest))
```

We will check that the predicted group and the original group are from the same distribution using the chi-squared test.

```python
from pyspark.ml.stat import ChiSquareTest
from pyspark.sql.functions import col
from scipy.stats import chi2_contingency

# create two sample dataframes
df1 = predictions_arrest.select("prediction").toPandas()
df2 = testData.select("label").toPandas()


# create a contingency table
contingency_table = pd.crosstab(df1['prediction'], df2['label'])

# perform chi-square test
stat, p, dof, expected = chi2_contingency(contingency_table)

# print the results of the chi-square test
print("Chi-square test results:")
print(f"Test statistic: {stat}")
print(f"P-value: {p}")
print(f"Degrees of freedom: {dof}")
print("Expected frequencies:")
print(expected)
```

The p-value of the chi-square test between the two groups is 1. This suggests that the two groups have similar distributions. It means there is no evidence to reject the null hypothesis that the two groups have the same distribution.

In other words, a p-value of 1 indicates that the observed differences between the two groups are likely to have occurred by chance alone. This suggests that the groups are similar concerning the variable being tested and that the null hypothesis cannot be rejected.

Therefore, a p-value of 1 indicates no significant difference in the distributions of the variable being tested between the two groups and that they are considered similar.

## 2.2.2 Classifying and Predicting Type of Crime

After the exploratory analysis, we already know that some primary crime types occur much more than others. To address that in our prediction and to prevent the prediction model from being biased, We will use oversampling method to balance the primary types distribution.

Since we have 25 Primary types, we will find the "Elbow" point in the Crime count we made during the EDS, and split the crimes in to minority and majority groups where the minority is the group of crimes which were from lower order of magnitude than the ones in the majority group.
Afterward, we will predict the primary type of crime based on its location and time orientation. 

```python
from pyspark.sql.functions import rand

# load the data with only the relevant columns
df_type = CrimesDataset.select("Beat","District","Ward","Month_Of_Year","Day_Of_Week","Week_of_Year", "Primary Type")

# limit the model to 150 thousand records to improve code runtime
#df_type = df_type.limit(150000)

# select based on the EDA the crime types which are the majority of the data  - for resampling process
Majority_of_crimes = ["THEFT","BATTERY","CRIMINAL DAMAGE","NARCOTICS","ROBBARY or TRESPASS","ASSAULT","OTHER OFFENSE","BURGLARY"\
                     ,"DECEPTIVE PRACTICE"]

# split the data into majority and minority by types
majority_class = df_type.filter(col("Primary Type").isin(Majority_of_crimes))
minority_class = df_type.filter(~col("Primary Type").isin(Majority_of_crimes))

# calculate the ratio of the minority class to the majority class
minority_count = minority_class.count()
print("minority_count: " ,minority_count)
majority_count = majority_class.count()
print("majority_count: ", majority_count)

# calculate the ratio of the minority class to the majority class
ratio = float(minority_count) / float(majority_count)
print("ratio: ",ratio)

# perform oversampling by sampling the minority class with replacement
oversampled_minority = minority_class.sampleBy('Primary Type', fractions={1: ratio}, seed=5544)
oversampled_data = majority_class.unionAll(oversampled_minority).orderBy(rand())

# asset the oversampled data back to the dataframe object
df_type = oversampled_data

#"Block","Beat","District","Ward","Day_Of_Week", "Month_Of_Year", "Longitude","Latitude", "Location Description"
vector_assembler = VectorAssembler(inputCols = ["Beat","District","Ward","Month_Of_Year"], outputCol="features")
df_temp = vector_assembler.transform(df_type)
featureIndexer =\
    VectorIndexer(inputCol="features", outputCol="indexedFeatures").fit(df_temp)

# index the labels 
labelIndexer = StringIndexer(inputCol="Primary Type", outputCol="indexedLabel").fit(df_temp)
labelConverter = IndexToString(inputCol="prediction", outputCol="predictedLabel",
                               labels=labelIndexer.labels)

# split the data into train and test sets
(trainingData, testData) = df_temp.randomSplit([0.7, 0.3])

# define the rf model
rf = RandomForestClassifier(labelCol="indexedLabel", featuresCol="indexedFeatures", numTrees=100,impurity='gini', maxBins=128)

pipeline = Pipeline(stages=[labelIndexer, featureIndexer, rf,labelConverter])

# fit the model to the train data
model = pipeline.fit(trainingData)

# test the model and print the results
predictions_type = model.transform(testData)
predictions_type.select("predictedLabel", "Primary Type").show()
print("Figure 8 : Random Forest results for Primary Type prediction")

# evaluate the results accuracy
evaluator = MulticlassClassificationEvaluator(
    labelCol="indexedLabel", predictionCol="prediction", metricName="accuracy")

# evaluate the results f1 score
evaluator2 = MulticlassClassificationEvaluator(
    labelCol="indexedLabel", predictionCol="prediction", metricName="f1")


accuracy = evaluator.evaluate(predictions_type)
f1 = evaluator2.evaluate(predictions_type)


print("accuracy = %g" % (accuracy))
print("f1 score = %g" % (f1))

print("Figure 9 : Random Forest results for Primary Type prediction grouped by predictedLabel")
predictions_type.groupBy("predictedLabel").count().show()
```

Using the chi-squared test, we will check that the predicted group and the original group are from the same distribution.

```python

# Create two sample dataframes
df1_type = predictions_type.select("predictedLabel").toPandas()
df2_type = testData.select("Primary Type").toPandas()


# create a contingency table
contingency_table = pd.crosstab(df1_type['predictedLabel'], df2_type['Primary Type'])

# perform chi-square test
stat, p, dof, expected = chi2_contingency(contingency_table)

# print the test results
print("Chi-square test results:")
print(f"Test statistic: {stat}")
print(f"P-value: {p}")
print(f"Degrees of freedom: {dof}")
print("Expected frequencies:")
print(expected)
```

A small p-value (less than 0.05) indicates that the observed differences between the groups are unlikely to have occurred by chance alone. Therefore, there is a significant difference between the two groups' distributions. 
Therefore, there is a significant difference in the distributions of the groups, and they cannot be considered similar.

## 2.2.3 Arrests Per Month Prediction

The following code generates a table with the number of arrests per each month of the dataset's five years.

```python
# Filter the dataset to include only the rows where Arrest=True
arrests_df = CrimesDataset.filter(CrimesDataset.Arrest == 1)

# Group the data by year and month, and sum the number of arrests for each group
result = arrests_df.groupBy('Year', 'Month_Of_Year').agg(sum('Arrest').alias('TotalArrests'))

# Show the average table

print("Figure 10 : Average arrests per month in chronological order")
result.orderBy("Year","Month_Of_Year").show()
```

```python
# Import necessary libraries


from statsmodels.tsa.statespace.sarimax import SARIMAX
from datetime import datetime
%matplotlib inline
from matplotlib.pylab import rcParams

from statsmodels.tsa.stattools import adfuller

# Aggregate the average crime count by month and year
agg_df = result.groupBy("Year","Month_Of_Year")\
                 .agg({"TotalArrests": "mean"})\
                 .orderBy("Year", "Month_Of_Year")

# Convert the PySpark DataFrame to Pandas DataFrame for time series analysis
pandas_df = agg_df.toPandas()

# Train a SARIMAX model on the data
model = SARIMAX(pandas_df["avg(TotalArrests)"], order=(3, 3, 6))
fitted_model = model.fit()

# Forecast the next 6 months average crime count
forecast = fitted_model.forecast(steps=6)

print("Figure 11 : SARIMAX model results for average arrests in the next six months prediction")

# Print the forecasted values
print(forecast)
```

```python
plt.figure(figsize=(15,7))
plt.plot(pandas_df["avg(TotalArrests)"], color='#379BDB', label='Original')
plt.plot(forecast, color='#D22A0D', label='Prediction')
plt.legend(loc='best')
plt.title('Figure 12: Original and predicted data of Arrest count')
plt.show(block=False)
```

```python

#Augmented Dickeyâ€“Fuller test:
print('Results of Dickey Fuller Test:')
dftest = adfuller(pandas_df["avg(TotalArrests)"], autolag='AIC')

dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
for key,value in dftest[4].items():
    dfoutput['Critical Value (%s)'%key] = value
    
print(dfoutput)
```

```python
#Augmented Dickeyâ€“Fuller test:
print('Results of Dickey Fuller Test:')
dftest = adfuller(forecast, autolag='AIC')

dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
for key,value in dftest[4].items():
    dfoutput['Critical Value (%s)'%key] = value
    
print(dfoutput)
```
